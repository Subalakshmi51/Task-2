```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the dataset
file_path = '/content/gretel_generated_table_2024-09-09-10-27-31.csv'  # Update this with your file path
data = pd.read_csv(file_path)

# Step 1: Display the dataset example
def view_dataset():
    print("Here is a preview of the dataset:")
    print(data.head())

# Step 2: Get details of a specific Iris sample by Sample ID
def display_sample_info(sample_id):
    # Retrieve the sample based on the Sample ID
    sample = data[data['sample id'] == sample_id]

    if sample.empty:
        print(f"Sample ID {sample_id} not found.")
        return

    print(f"\nDetails of Sample ID {sample_id}:")
    print(sample[['Sepal length (cm)', 'Sepal width (cm)', 'Petal length (cm)', 'Petal width (cm)', 'Class', 'Country']])

    # Visualize the sample on a scatter plot
    plt.figure(figsize=(8,6))
    sns.scatterplot(x='Sepal length (cm)', y='Petal length (cm)', hue='Class', data=data, palette='Set1', s=100)

    # Highlight the specific sample
    plt.scatter(sample['Sepal length (cm)'], sample['Petal length (cm)'], color='red', s=200, label=f'Sample ID: {sample_id}', edgecolor='black')

    plt.title(f'Sepal Length vs Petal Length (Sample ID: {sample_id})')
    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Petal Length (cm)')
    plt.legend()
    plt.show()

# Step 3: Display which countries contain the specified Iris class
def display_class_countries(iris_class):
    countries = data[data['Class'] == iris_class]['Country'].unique()

    if len(countries) == 0:
        print(f"No countries found for Iris class: {iris_class}")
    else:
        print(f"Iris class {iris_class} is found in the following countries:")
        for country in countries:
            print(country)

# Step 4: Pie chart visualization of lengths and widths by Iris class
def plot_pie_chart():
    fig, ax = plt.subplots(2, 2, figsize=(10, 8))

    data.groupby('Class')['Sepal length (cm)'].sum().plot.pie(autopct='%1.1f%%', ax=ax[0, 0], title='Sepal Length')
    data.groupby('Class')['Sepal width (cm)'].sum().plot.pie(autopct='%1.1f%%', ax=ax[0, 1], title='Sepal Width')
    data.groupby('Class')['Petal length (cm)'].sum().plot.pie(autopct='%1.1f%%', ax=ax[1, 0], title='Petal Length')
    data.groupby('Class')['Petal width (cm)'].sum().plot.pie(autopct='%1.1f%%', ax=ax[1, 1], title='Petal Width')

    plt.tight_layout()
    plt.show()

# Step 5: World map visualization showing all Iris species by country
def display_world_map():
    # Aggregate data by country and class to show all occurrences of Iris species in each country
    country_iris = data.groupby(['Country', 'Class']).size().reset_index(name='Count')

    # Create world map visualization using Plotly
    fig = px.choropleth(
        country_iris,
        locations="Country",
        locationmode="country names",
        color="Class",
        hover_name="Country",
        hover_data=["Count"],
        title="Iris Species Distribution by Country",
        color_discrete_map={
            'Iris-setosa': 'blue',
            'Iris-versicolor': 'green',
            'Iris-virginica': 'red'
        }
    )

    fig.update_layout(geo=dict(showframe=False, showcoastlines=False, projection_type='equirectangular'))
    fig.show()

# Main Menu
while True:
    print("\nChoose an option:")
    print("1. View a preview of the dataset")
    print("2. View a specific sample by Sample ID")
    print("3. View countries with a specific Iris class")
    print("4. View pie charts of lengths and widths by Iris class")
    print("5. View World Map of Iris Species by Country")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        view_dataset()
    elif choice == '2':
        sample_id = input("Enter the Sample ID: ")
        display_sample_info(sample_id)
    elif choice == '3':
        iris_class = input("Enter the Iris class (e.g., Iris-setosa, Iris-versicolor, Iris-virginica): ")
        display_class_countries(iris_class)
    elif choice == '4':
        plot_pie_chart()
    elif choice == '5':
        display_world_map()
    elif choice == '6':
        print("Exiting...")
        break
    else:
        print("Invalid choice. Please try again.")
```
